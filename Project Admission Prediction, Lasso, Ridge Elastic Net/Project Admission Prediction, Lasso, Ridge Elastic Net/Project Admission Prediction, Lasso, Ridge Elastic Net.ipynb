#### accuracy

```python
lr.score(x_test,y_test) # This is R^2.
```

```python
# Lets create a function to create adjected-R^2
def adj_r2(x,y):
    r2=lr.score(x,y)
    n=x.shape[0]
    p=x.shape[1]
    adjected_r2=1-(1-r2)*(n-1)/(n-p-1)
    return adjected_r2
```

```python
adj_r2(x_test,y_test)
```
